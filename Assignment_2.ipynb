Importing Libraries and Data Set for Analysis

```python
import pandas as pd
df = pd.read_csv("summerdata.csv")
```

**1) In how many cities, Summer Olympics is held so far ?**

```python
print(len(df['City'].unique()))
```

**2) Show the Bar Graph of Top 5 Sport having most number of Gold Medals till now ?**

```python
data = []
gold = df[df['Medal']== 'Gold']
for sport in gold['Sport'].unique():
    data.append([sport,len(gold[gold['Sport']==sport])])
data = pd.DataFrame(data, columns= ['Sports','Gold Medals'])
data = data.sort_values(by = 'Gold Medals', ascending = False).head()
data.plot(x = 'Sports', y = 'Gold Medals', kind = 'bar', figsize = (10,5))    
```

**3) Show me the bar chart of Top 5 sport having most number of medals till now?**

```python
data = []
for sport in df['Sport'].unique():
  data.append([sport,len(df[df['Sport'] == sport])])
data = pd.DataFrame(data, columns= ['Sports','Medals'])
data = data.sort_values(by = 'Medals', ascending = False).head()
data.plot(x = 'Sports', y = 'Medals', kind = 'bar', figsize = (10,5))
```

**4) Show me the  bar chart of top 5 player who has won most number of medals till now?**

```python
data = []
for Athlete in df['Athlete'].unique():
  data.append([Athlete,len(df[df['Athlete'] == Athlete])])
data = pd.DataFrame(data, columns= ['Athlete','Medals'])
data = data.sort_values(by = 'Medals', ascending = False).head()
data.plot(x = 'Athlete', y = 'Medals', kind = 'bar', figsize = (10,5))
```

**5) Show me the bar chart of top 5 player who has won most number Gold Medals of medals till now ?**

```python
data = []
gold = df[df['Medal'] == 'Gold']
for Athlete in gold['Athlete'].unique():
  data.append([Athlete,len(gold[gold['Athlete'] == Athlete])])
data = pd.DataFrame(data, columns= ['Athlete','Gold Medals'])
data = data.sort_values(by = 'Gold Medals', ascending = False).head()
data.plot(x = 'Athlete', y = 'Gold Medals', kind = 'bar', figsize = (10,5))
```

**6) Show me in which year India won first Gold Medal in Summer Olympics?**

```python
IND = df[df['Country'] == 'IND']
gold = IND[IND['Medal'] == 'Gold']
print(gold.min()['Year'])
```

**7) Show me the bar chart of top 5 event which are most popular in terms on number of players ?**

```python
data = []
for Event in df['Event'].unique():
  evt = df[df['Event'] == Event]
  data.append([Event,len(evt['Athlete'].unique())])
data = pd.DataFrame(data, columns= ['Event','No. of Players'])
data = data.sort_values(by = 'No. of Players', ascending = False).head()
data.plot(x = 'Event', y = 'No. of Players', kind = 'bar', figsize = (10,5))
```

**8) Show me the bar chart of top 5 sport having most female Gold Medalists ?**

```python
data = []
female = df[df['Gender'] == 'Women']
gold = female[female['Medal'] == 'Gold']
for sport in gold['Sport'].unique():
  data.append([sport,len(gold[gold['Sport'] == sport])])
data = pd.DataFrame(data, columns= ['Sports','Gold Medals'])
data = data.sort_values(by = 'Gold Medals', ascending = False).head()
data.plot(x = 'Sports', y = 'Gold Medals', kind = 'bar', figsize = (10,5))
```
